```python
# 선물 과거 데이터 조회

import ccxt 
import pprint
import time
from datetime import datetime
import pandas as pd
import numpy as np
import talib
import math

with open("api_test1.txt") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret  = lines[1].strip()

binance = ccxt.binance(config={
    'apiKey': api_key, 
    'secret': secret,
    'enableRateLimit': True,
    'options': {
        'defaultType': 'future'
    }
})



while True:
    now = datetime.now()

    sol = binance.fetch_ohlcv(
        symbol = "SOL/USDT",
        timeframe='1m',
        since=None,
        limit=1000
    )

    df = pd.DataFrame(sol, columns=['datetime', 'open', 'high', 'low', 'close', 'volume'])
    df['datetime'] = pd.to_datetime(df['datetime'], unit='ms')
    df.set_index('datetime', inplace=True)


    # ma60
    df['ma60'] = talib.MA(np.asarray(df['close']), 60)
    df['ma60_incli'] = df.ma60 - df.ma60.shift()
    df['ma60_degrees'] = df['ma60_incli'].apply(lambda x: math.degrees(math.atan2(x, 3*60))) 
    df['ma60_minus_range'] = df.apply(lambda x: 0.5* ((x['ma60'] - x['close'])**2) * math.tan(math.radians(x['ma60_degrees']))/60, axis=1)
    df['ma60_mrma'] = talib.SMA(np.asarray(df['ma60_minus_range']), 60)
    df['ma60_mrinc'] = df.ma60_mrma - df.ma60_mrma.shift()
    df['ma60_mrdeg'] = df['ma60_mrinc'].apply(lambda x: math.degrees(math.atan2(x, 3*60)))


    # ma100
    df['ma100'] = talib.MA(np.asarray(df['close']), 100)
    df['ma100_incli'] = df.ma100 - df.ma100.shift()
    df['ma100_degrees'] = df['ma100_incli'].apply(lambda x: math.degrees(math.atan2(x, 3*60))) 
    df['ma100_minus_range'] = df.apply(lambda x: 0.5* ((x['ma100'] - x['close'])**2) * math.tan(math.radians(x['ma100_degrees']))/100, axis=1)
    df['ma100_mrma'] = talib.SMA(np.asarray(df['ma100_minus_range']), 60)
    df['ma100_mrinc'] = df.ma100_mrma - df.ma100_mrma.shift()
    df['ma100_mrdeg'] = df['ma100_mrinc'].apply(lambda x: math.degrees(math.atan2(x, 3*60)))

    # ma120
    df['ma120'] = talib.MA(np.asarray(df['close']), 120)
    df['ma120_incli'] = df.ma120 - df.ma120.shift()
    df['ma120_degrees'] = df['ma120_incli'].apply(lambda x: math.degrees(math.atan2(x, 3*60))) 
    df['ma120_minus_range'] = df.apply(lambda x: 0.5* ((x['ma120'] - x['close'])**2) * math.tan(math.radians(x['ma120_degrees']))/120, axis=1)
    df['ma120_mrma'] = talib.SMA(np.asarray(df['ma120_minus_range']), 60)
    df['ma120_mrinc'] = df.ma120_mrma - df.ma120_mrma.shift()
    df['ma120_mrdeg'] = df['ma120_mrinc'].apply(lambda x: math.degrees(math.atan2(x, 3*60)))

    # ma200
    df['ma200'] = talib.MA(np.asarray(df['close']), 200)
    df['ma200_incli'] = df.ma200 - df.ma200.shift()
    df['ma200_degrees'] = df['ma200_incli'].apply(lambda x: math.degrees(math.atan2(x, 3*60))) 
    df['ma200_minus_range'] = df.apply(lambda x: 0.5* ((x['ma200'] - x['close'])**2) * math.tan(math.radians(x['ma200_degrees']))/200, axis=1)
    df['ma200_mrma'] = talib.SMA(np.asarray(df['ma200_minus_range']), 60)
    df['ma200_mrinc'] = df.ma200_mrma - df.ma200_mrma.shift()
    df['ma200_mrdeg'] = df['ma200_mrinc'].apply(lambda x: math.degrees(math.atan2(x, 3*60)))

    # ma240
    df['ma240'] = talib.MA(np.asarray(df['close']), 240)
    df['ma240_incli'] = df.ma240 - df.ma240.shift()
    df['ma240_degrees'] = df['ma240_incli'].apply(lambda x: math.degrees(math.atan2(x, 3*60))) 
    df['ma240_minus_range'] = df.apply(lambda x: 0.5* ((x['ma240'] - x['close'])**2) * math.tan(math.radians(x['ma240_degrees']))/240, axis=1)
    df['ma240_mrma'] = talib.SMA(np.asarray(df['ma240_minus_range']), 60)
    df['ma240_mrinc'] = df.ma240_mrma - df.ma240_mrma.shift()
    df['ma240_mrdeg'] = df['ma240_mrinc'].apply(lambda x: math.degrees(math.atan2(x, 3*60)))

    print(now.strftime('%Y-%m-%d %H:%M:%S'))
    print('mrdeg')
    print('ma60 : ', df['ma60_mrdeg'][-1])
    print('ma100 : ', df['ma100_mrdeg'][-1])
    print('ma120 : ', df['ma120_mrdeg'][-1])
    print('ma200 : ', df['ma200_mrdeg'][-1])
    print('ma240 : ', df['ma240_mrdeg'][-1])
    print()
    print('degrees')
    print('ma60 : ', df['ma60_degrees'][-1])
    print('ma100 : ', df['ma100_degrees'][-1])
    print('ma120 : ', df['ma120_degrees'][-1])
    print('ma200 : ', df['ma200_degrees'][-1])
    print('ma240 : ', df['ma240_degrees'][-1])
    print()
    print('-------------------------------------------')
    print()

    time.sleep(60)
```
