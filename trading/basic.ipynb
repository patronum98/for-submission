```python
# 시장가 주문과 TP/SL

import ccxt 
import pprint

with open(".txt") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret  = lines[1].strip()

binance = ccxt.binance(config={
    'apiKey': api_key, 
    'secret': secret,
    'enableRateLimit': True,
    'options': {
        'defaultType': 'future'
    }
})

orders = [None] * 3

# market price
orders[0] = binance.create_order(
    symbol = 'SOL/USDT',
    type='MARKET',
    side='buy',
    amount=1
)

# take profit
orders[1] = binance.create_order(
    symbol="SOL/USDT",
    type="TAKE_PROFIT_MARKET",
    side="sell",
    amount=2,
    params={'stopPrice': 34.7}
)

# stop loss
orders[2] = binance.create_order(
    symbol="SOL/USDT",
    type="STOP_MARKET",
    side="sell",
    amount=2,
    params={'stopPrice': 34.46}
)

for order in orders:
    pprint.pprint(order)
```

```python
# 지정가 주문과 TP/SL

import ccxt 
import pprint

with open(".txt") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret  = lines[1].strip()

binance = ccxt.binance(config={
    'apiKey': api_key, 
    'secret': secret,
    'enableRateLimit': True,
    'options': {
        'defaultType': 'future'
    }
})

orders = [None] * 3
price = 27

# limit price
orders[0] = binance.create_order(
    symbol='SOL/USDT',
    type='LIMIT',
    side='buy',
    amount=1,
    price=price
)

# take profit
orders[1] = binance.create_order(
    symbol="SOL/USDT",
    type="TAKE_PROFIT",
    side="sell",
    amount=1,
    price=price,
    params={'stopPrice': 50}
)

# stop loss
orders[2] = binance.create_order(
    symbol="SOL/USDT",
    type="STOP",
    side="sell",
    amount=1,
    price=price,
    params={'stopPrice': 25}
)

for order in orders:
    pprint.pprint(order)
```

```python
# 선물 현재가 출력하기

import ccxt
import time

with open("api_test1.txt") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret  = lines[1].strip()

binance = ccxt.binance(config={
    'apiKey': api_key, 
    'secret': secret,
    'enableRateLimit': True,
    'options': {
        'defaultType': 'future'
    }
})

symbol='SOL/USDT'

while True:
    sol = binance.fetch_ticker(symbol)
    print(sol['last'])
    time.sleep(1)
```

```python
# 현재시간 출력하기

import ccxt
import time
import datetime

with open("api_test1.txt") as f:
    lines = f.readlines()
    api_key = lines[0].strip()
    secret  = lines[1].strip()

binance = ccxt.binance(config={
    'apiKey': api_key, 
    'secret': secret,
    'enableRateLimit': True,
    'options': {
        'defaultType': 'future'
    }
})

symbol='SOL/USDT'

while True:
    sol = binance.fetch_ticker(symbol)
    now = datetime.datetime.now()
    print(now, sol['last'])
    time.sleep(1)
```
